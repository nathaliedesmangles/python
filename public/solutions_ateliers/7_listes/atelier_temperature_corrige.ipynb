
# üî¨ Analyse des temp√©ratures d'une semaine (corrig√©, sans compr√©hension de liste ni CSV)

Ce notebook travaille avec une liste de cha√Ænes d√©j√† en m√©moire. On extrait, analyse et visualise les donn√©es m√©t√©o.


```python

# Donn√©es fournies sous forme de cha√Ænes
donnees = [
    "12.3, 16.8, 14.0",
    "11.5, 18.2, 15.4",
    "10.8, 17.6, 14.9",
    "13.0, 19.1, 16.3",
    "14.1, 20.2, 18.5",
    "12.9, 18.7, 16.2",
    "11.7, 17.8, 15.0"
]
```

```python

# Transformation des cha√Ænes en listes de 3 nombres flottants
temperatures = []
for ligne in donnees:
    morceaux = ligne.split(",")
    valeurs = [float(x.strip()) for x in morceaux]
    temperatures.append(valeurs)

print("Liste des temp√©ratures (liste imbriqu√©e) :")
print(temperatures)
```

```python

# Moyenne de chaque jour
moyennes_journali√®res = []
for jour in temperatures:
    moyenne = sum(jour) / 3
    moyennes_journali√®res.append(moyenne)

print("Temp√©ratures moyennes par jour :")
print(moyennes_journali√®res)

# Temp√©rature maximale
valeurs_toutes = []
for jour in temperatures:
    for t in jour:
        valeurs_toutes.append(t)

temp_max = max(valeurs_toutes)
jour_max = valeurs_toutes.index(temp_max) // 3 + 1

print(f"Temp√©rature maximale : {temp_max}¬∞C (Jour {jour_max})")

# Moyenne g√©n√©rale
moyenne_globale = sum(valeurs_toutes) / len(valeurs_toutes)
print(f"Temp√©rature moyenne de la semaine : {moyenne_globale:.2f}¬∞C")
```

```python

import matplotlib.pyplot as plt

jours = [1, 2, 3, 4, 5, 6, 7]

plt.plot(jours, moyennes_journali√®res, marker='o', color='blue', label='Moyenne journali√®re')
plt.title("Temp√©ratures moyennes par jour")
plt.xlabel("Jour")
plt.ylabel("Temp√©rature (¬∞C)")
plt.grid()
plt.legend()
plt.savefig("graphique_temperature.png")
plt.show()
```

```python

# Recompte des jours o√π la temp√©rature de midi d√©passe 18¬∞C
nb_midi_haut = 0

for ligne in donnees:
    morceaux = ligne.split(",")
    midi = float(morceaux[1].strip())
    if midi > 18:
        nb_midi_haut += 1

print(f"Nombre de jours o√π la temp√©rature de midi a d√©pass√© 18¬∞C : {nb_midi_haut}")
```
